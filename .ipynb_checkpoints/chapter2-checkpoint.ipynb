```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import warnings
import seaborn as sns
import pandas as pd
import matplotlib as mpl
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
def f(x):
    return x * np.sin(x)
```

```python
DIM = 200
SAMPLE_DIM = 30
DEGREES = [1,2,5,8,11,16,19,22]
X_MAX = 30

x_plot = np.linspace(0, 10, DIM)
x = np.linspace(0, 10, SAMPLE_DIM)
y_plus_noise = f(x) + np.random.normal(size=SAMPLE_DIM)

X = x[:, np.newaxis]
y = y_plus_noise[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]
y_plot = f(X_plot)

colors = sns.color_palette("hls", len(DEGREES))
lw = 2

plt.figure(figsize=(20,10))
plt.plot(X_plot, y_plot, color='black', linewidth=lw*2,label="ground truth")
plt.scatter(x, y, color='navy', s=30, marker='o', label="training points")

mse_train = []
mse_test = []

for count, degree in enumerate(DEGREES):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X,y)
    y_plot = model.predict(X_plot)
    plt.plot(x_plot, y_plot, color=colors[count], linewidth=lw,label="degree %d" % degree)

plt.legend(loc='lower left')

plt.show()
```

```python
DIM = 200
SAMPLE_DIM = 50
DEGREES = 30

x_plot = np.linspace(0, 10, DIM)
x = np.linspace(0, 10, SAMPLE_DIM)
y_plus_noise = f(x) + np.random.normal(size=SAMPLE_DIM)

X_train = x[:, np.newaxis]
y_train = y_plus_noise[:, np.newaxis]

mse_train = []
mse_test = []

for degree in range(1,DEGREES): 
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_train,y_train)
    sample_indices = np.random.choice(DIM, SAMPLE_DIM,replace=False)
    X_test = x_plot[sample_indices].reshape(-1,1)
    y_test = y_plot[sample_indices].reshape(-1,1)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    mse_train.append(mean_squared_error(y_train, y_train_pred))
    mse_test.append(mean_squared_error(y_test, y_test_pred))
    
plt.figure(figsize=(15,7))
plt.plot([*range(1,DEGREES)],mse_train, color='C1', label = 'Train mse')
plt.plot([*range(1,DEGREES)],mse_test, color='C2', label = 'Test mse')
plt.legend(loc='lower left')
plt.show()
    
```
